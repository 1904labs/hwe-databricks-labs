# Week 3 Demo: Writing Your First Tests

This notebook contains two simple SQL queries used to demonstrate how pytest tests work.

Each query is tagged with `-- @test:tag_name` so the test file can find and run it.

## Query 1: Find Engineering Employees

Returns all employees in the Engineering department.
We will use `len()` to assert the correct number of rows are returned.

```sql
-- @test:demo_engineering_employees
SELECT *
FROM week3_testing.employees
WHERE department = 'Engineering'
```

## Query 2: Find a Specific Employee

Returns a single employee by ID.
We will assert that the `name` column equals the expected value.

```sql
-- @test:demo_find_employee
SELECT *
FROM week3_testing.employees
WHERE employee_id = 'EMP-001'
```
